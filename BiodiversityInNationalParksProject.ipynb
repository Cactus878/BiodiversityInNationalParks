```python
observations = pd.read_csv("observations.csv")
observations
```

```python
observations.describe()
```

```python
observations['park_name'].value_counts().plot(kind='bar', color='skyblue')

# Customize the chart
plt.title('Category Frequency Distribution')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
#Combine both data tables where rows of data from both datasets own the same scientific_name
merged_data = pd.merge(observations, species_info, on='scientific_name')
merged_data
```

```python
#Create all possibilities between category and park_name. Additionnaly add up all together how many times a category is seen in a specific park.
grouped_data = merged_data.groupby(['park_name', 'category']).agg({'observations': 'sum'}).reset_index()
grouped_data
```

```python
#Improve visuals
pivot_data = grouped_data.pivot(index='park_name', columns='category', values='observations').fillna(0)
pivot_data
```

```python
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 10))

plt.title('Number of Animal Observations by Class in Each Park')
plt.xlabel('Park Name')
plt.ylabel('Number of Observations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Animal Class')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
grouped_data = merged_data.groupby(['conservation_status', 'category']).agg({'observations': 'sum'}).reset_index()
grouped_data
```

```python
pivot_data = grouped_data.pivot(index='conservation_status', columns='category', values='observations').fillna(0)
pivot_data
```

```python
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 10))

plt.title('Number of Animal Observations by Conservation')
plt.xlabel('Conservations')
plt.ylabel('Number of Observations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Animal Class')
plt.tight_layout()

# Show the plot
plt.show()
```
